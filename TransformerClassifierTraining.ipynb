## risk labeling

```python
import re

def detect_risk(text):
    text = text.lower()

    # Suicide intent keywords
    suicide_kw = [
        r"kill myself", r"end my life", r"suicid", r"i don't want to live",
        r"i can’t go on", r"i cant go on", r"take my life"
    ]

    # Self-harm
    selfharm_kw = [
        r"cut myself", r"self harm", r"hurt myself", r"cutting again"
    ]

    # Violence
    violence_kw = [
        r"kill him", r"hurt someone", r"hurt her", r"i want to hurt"
    ]

    # Crisis signals
    crisis_kw = [
        r"no point in living", r"nothing matters", r"give up", r"end everything"
    ]

    kws = suicide_kw + selfharm_kw + violence_kw + crisis_kw

    for kw in kws:
        if re.search(kw, text):
            return 1

    return 0

df_final["risk_label"] = df_final["input"].astype(str).apply(detect_risk)
```

```python
# ======================================
#  SAFE & COMPATIBLE CLASSIFIER TRAINING
# ======================================

import os
import shutil
import numpy as np
from datasets import Dataset
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
from sklearn.model_selection import train_test_split

from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
    TextClassificationPipeline,
)

os.environ["WANDB_DISABLED"] = "true"

# ===============================
# 1. PREPARE DATA
# ===============================
clf_df = df_final[["input", "risk_label"]].dropna().reset_index(drop=True)
clf_df = clf_df.rename(columns={"risk_label": "label"})

print("Classifier dataset size:", clf_df.shape)

train_df, eval_df = train_test_split(
    clf_df,
    test_size=0.1,
    stratify=clf_df["label"],
    random_state=42,
)

train_ds = Dataset.from_pandas(train_df)
eval_ds = Dataset.from_pandas(eval_df)

# ===============================
# 2. MODEL + TOKENIZER
# ===============================
model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=2
)

# ===============================
# 3. TOKENIZATION
# ===============================
def tokenize_fn(batch):
    return tokenizer(
        batch["input"],
        truncation=True,
        padding="max_length",
        max_length=256
    )

train_tok = train_ds.map(tokenize_fn, batched=True)
eval_tok = eval_ds.map(tokenize_fn, batched=True)

for col in ["input", "__index_level_0__"]:
    if col in train_tok.column_names:
        train_tok = train_tok.remove_columns(col)
    if col in eval_tok.column_names:
        eval_tok = eval_tok.remove_columns(col)

train_tok = train_tok.with_format("torch")
eval_tok = eval_tok.with_format("torch")

# ===============================
# 4. TRAINING ARGUMENTS
# ===============================
training_args = TrainingArguments(
    output_dir="./sentiment_risk_classifier",
    save_strategy="no",      # <- safest option for old HF versions
    num_train_epochs=3,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=32,
    learning_rate=2e-5,
    weight_decay=0.01,
    logging_steps=100,
    report_to="none",
)

# ===============================
# 5. METRICS
# ===============================
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    p, r, f1, _ = precision_recall_fscore_support(labels, preds, average="binary")
    acc = accuracy_score(labels, preds)
    return {"accuracy": acc, "precision": p, "recall": r, "f1": f1}

# ===============================
# 6. TRAINER
# ===============================
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_tok,
    eval_dataset=eval_tok,
    compute_metrics=compute_metrics,
)

# ===============================
# 7. TRAIN
# ===============================
trainer.train()

# ===============================
# 8. EVALUATE ONCE (SAFE)
# ===============================
metrics = trainer.evaluate()
print("\nFinal Evaluation Metrics:")
print(metrics)

best_f1 = metrics["eval_f1"]

# ===============================
# 9. SAVE BEST MODEL MANUALLY
# ===============================
BEST_PATH = "./best_risk_classifier"

if os.path.exists(BEST_PATH):
    shutil.rmtree(BEST_PATH)

trainer.model.save_pretrained(BEST_PATH)
tokenizer.save_pretrained(BEST_PATH)

print("\nBest model saved to:", BEST_PATH)

# ===============================
# 10. INFERENCE
# ===============================
best_model = AutoModelForSequenceClassification.from_pretrained(BEST_PATH)
best_tokenizer = AutoTokenizer.from_pretrained(BEST_PATH)

pipeline = TextClassificationPipeline(
    model=best_model,
    tokenizer=best_tokenizer,
    task="text-classification",
    top_k=1
)

def is_risky(text, threshold=0.5):
    result = pipeline(text)[0]
    return result["label"] == "LABEL_1" and result["score"] >= threshold


```

```python

import pandas as pd

# 1. Weighted lexicon for suicide / distress risk
risk_lexicon = {
    # ===== HIGH RISK (weight 3) =====
    "kill myself": 3,
    "killing myself": 3,
    "want to die": 3,
    "end my life": 3,
    "suicide": 3,
    "take my life": 3,
    "cut myself": 3,
    "hurt myself": 3,
    "ending it all": 3,

    # ===== MODERATE RISK (weight 2) =====
    "wish i could disappear": 2,
    "i don't want to be here": 2,
    "better if i was gone": 2,
    "can't go on": 2,
    "life is not worth living": 2,
    "no point in living": 2,
    "i'm afraid of what i might do": 2,

    # ===== MILD DISTRESS (weight 1) =====
    "empty": 1,
    "worthless": 1,
    "numb": 1,
    "hopeless": 1,
    "helpless": 1,
    "overwhelmed": 1,
    "i feel alone": 1,
    "i can't handle this": 1,
    "i'm struggling": 1,
}

# 2. Risk scoring function
def lexicon_risk_score(text, lexicon=risk_lexicon):
    text_lower = str(text).lower()
    score = 0

    # accumulate weights for all matching phrases
    for phrase, weight in lexicon.items():
        if phrase in text_lower:
            score += weight

    # map accumulated score to 0–3
    if score >= 3:
        return 3
    elif score == 2:
        return 2
    elif score == 1:
        return 1
    else:
        return 0

# 3. Apply to your dataframe
df_final["risk_label"] = df_final["input"].apply(lexicon_risk_score)

print(df_final["risk_label"].value_counts())
```

```python
def format_sample(example):
    prompt = (
        f"{example['instruction']}\n\n"
        f"User: {example['input']}\n\n"
        f"Assistant:"
    )

    tokenized = tokenizer(
        prompt,
        truncation=True,
        max_length=1024
    )

    labels = tokenizer(
        example["output"],
        truncation=True,
        max_length=1024
    )

    return {"input_ids": tokenized["input_ids"],
            "attention_mask": tokenized["attention_mask"],
            "labels": labels["input_ids"]}


```
